```python
from pydantic import BaseModel
```

```python
class Student(BaseModel):
    name:str
    age:int
    year:str
```

```python
student= Student(name='tom',age=20,year='kindergarden')
```

```python
for key ,value in student:
    print (student.dict()[key])
```

```python
Student(**student.dict())
```
